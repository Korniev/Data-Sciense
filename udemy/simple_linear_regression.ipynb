```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set_theme()

from sklearn.linear_model import LinearRegression
```

### Statistical prediction for Simple Linear Regression

```python
data = pd.read_csv('/1.01.+Simple+linear+regression.csv')
data.head()
```

```python
data.describe()
```

Create first regression

```python
y = data['GPA']
x1 = data['SAT']
```

```python
plt.scatter(x1, y)
plt.xlabel('SAT', fontsize=20)
plt.ylabel('GPA', fontsize=20)
plt.show()
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(x1,y)
y_hat = 0.0017 * x1 + 0.275
fig = plt.plot(x1, y_hat, lw=4, c='orange', label='regression line')
plt.xlabel('SAT', fontsize=20)
plt.ylabel('GPA', fontsize=20)
plt.show()
```

##sklearn Simple Linear Regression implementation

```python
data = pd.read_csv('real_estate_price_size.csv')
data.head()
```

```python
data.describe()
```

### Declare the dependent and the independent variables

```python
y = data['price']
x = data['size']
```

Explore the data:

```python
plt.scatter(x,y)
plt.xlabel('Size',fontsize=20)
plt.ylabel('Price',fontsize=20)
plt.show()
```

### Transform the inputs into a matrix (2D object)

```python
x_matrix = x.values.reshape(-1,1)
x_matrix[:5]
```

### Regression itself

```python
reg = LinearRegression()
reg.fit(x_matrix, y)
```

### Calculate the R-squared

```python
reg.score(x_matrix, y)
```

### Find the intercept

```python
reg.intercept_
```

### Find the coefficients

```python
reg.coef_
```

### Making predictions

You find an apartment online with a size of 750 sq.ft.

All else equal what should be its price according to the model?

```python
prediction = reg.predict(np.array([750]).reshape(-1, 1))
print(f'The price for 750 m2 apartment is {prediction}')
```
