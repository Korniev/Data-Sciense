```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
data = pd.read_csv('/1.01.+Simple+linear+regression.csv')
data.head()
```

```python
data.describe()
```

Create first regression

```python
y = data['GPA']
x1 = data['SAT']
```

```python
plt.scatter(x1, y)
plt.xlabel('SAT', fontsize=20)
plt.ylabel('GPA', fontsize=20)
plt.show()
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(x1,y)
y_hat = 0.0017 * x1 + 0.275
fig = plt.plot(x1, y_hat, lw=4, c='orange', label='regression line')
plt.xlabel('SAT', fontsize=20)
plt.ylabel('GPA', fontsize=20)
plt.show()
```

Task Simple Linear regression

```python
data_estate = pd.read_csv('/content/real_estate_price_size.csv')
data_estate.head()
```

```python
data_estate.describe()
```

```python
y = data_estate['price']
x1 = data_estate['size']
```

```python
plt.scatter(x1, y)
plt.xlabel('size', fontsize=20)
plt.ylabel('price', fontsize=20)
plt.show()
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(x1,y)
y_hat = 223.1787 * x1 + 1.019e+05
fig = plt.plot(x1, y_hat, lw=4, c='orange', label='regression line')
plt.xlabel('size', fontsize=20)
plt.ylabel('price', fontsize=20)
plt.show()
```
